```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
```

# 2. YOLO 인스턴스 생성

```python
model = YOLO('yolov8n.pt')
```

# 3. 이미지의 사물 인식처리

```python
results = model.predict('/content/drive/MyDrive/data/mission_impossible_fallout.jpg')

print(results)
```

# 4. 결과 시각화

```python
result_image = results[0].plot()

plt.imshow(cv2.cvtColor(result_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
