```python
def minmax(series):
    min = np.min(series)
    max = np.max(series)
    print("최소값은 {}, 최대값은 {}입니다.".format(min, max))
    return (series-min)/(max-min)
```

```python
normalized = minmax(df['거래금액'])

print("정규화 후 최소값은 {:.1f}, 최대값은 {:.1f}입니다.".format(np.min(normalized), np.max(normalized)))
print(normalized)
```

```python
def standard(series):
    std = np.std(series)
    mean = np.mean(series)
    print("표준편차는 {}, 평균은 {}입니다.".format(std, mean))
    return (series-mean)/std
```

```python
standardized = standard(df['거래금액'])

print("표준화 후 최소값은 {}, 최대값은 {}입니다.".format(np.min(standardized), np.max(standardized)))

print(standardized)
```
