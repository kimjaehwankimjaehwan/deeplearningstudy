```r
data <- c(7, 9, 6, 10, 12, 14, 9, 8)
boxplot(data)
```

```r
# 데이터프레임 생성
df <- data.frame(
  value = c(7, 9, 6, 10, 12, 14, 9, 8, 15, 13),
  group = c("A", "A", "A", "A", "A", "B", "B", "B", "B", "B")
)

# 그룹별 박스플롯
boxplot(value ~ group, data = df)
```

```r
boxplot(value ~ group, data = df,
        col = c("skyblue", "lightgreen"),
        main = "Group-wise Boxplot",
        xlab = "Group", ylab = "Value")
```
