```python
import pandas as pd
import numpy as np
```

```python
# CSV 파일 불러오기
df_raw = pd.read_csv("/content/drive/MyDrive/kdt_2506/data/kpop_idol_followers.csv")
```

```python
df_drop = df_raw[['Followers', 'Gender.x', 'Company', 'Country', 'Second.Country', 'Height', 'Weight', 'age', 'year.career']]
```

```python
df_drop.head()
```

```python
df = pd.get_dummies(df_drop)
```

```python
df.shape
```

```python
df_corr = pd.DataFrame({'corr':df.corr()['Followers']})
df_corr['corr_abs'] = df_corr['corr'].abs()
df_corr.sort_values(by='corr_abs', ascending=False, inplace=True)
df_corr.head(10)
```

```python
y = df['Followers']
X = df.drop(columns='Followers')
y.shape, X.shape
```

```python
# RandomForest 회귀분석 기본 코드
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=777)

# 모델 학습
model = RandomForestRegressor(random_state=777)
model.fit(X_train, y_train)

# 예측 및 평가
y_pred = model.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("RMSE:", rmse)
print("R²:", r2)
```

```python
from sklearn.model_selection import cross_val_score, cross_validate, KFold
from sklearn.ensemble import RandomForestRegressor
```

```python
cv = KFold(n_splits=3, shuffle=True, random_state=777)
model = RandomForestRegressor(random_state=777)
cv_results = cross_validate(
    model,
    X,
    y,
    cv=cv,
    scoring=['r2', 'neg_root_mean_squared_error'],
    return_train_score=True
)
```

```python
# R² 출력
r2_train = cv_results['train_r2']
r2_test = cv_results['test_r2']

print("=== R² (결정계수) ===")

for i, (score_train, score_test) in enumerate(zip(r2_train, r2_test)):
    print(f"Fold {i+1}: {score_train:.4f}, {score_test:.4f}")
print(f"평균 R²: {np.mean(r2_train):.4f}, {np.mean(r2_test):.4f}")

print()

# MSE 출력 (음수로 반환되므로 부호 반전)
rmse_scores = -cv_results['test_neg_root_mean_squared_error']
print("=== RMSE ===")
for i, score in enumerate(rmse_scores):
    print(f"Fold {i+1}: {score:.4f}")
print(f"평균 RMSE: {np.mean(rmse_scores):.4f}")
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.metrics import make_scorer, mean_squared_error
import numpy as np

# 결측치 처리
X_mean = X.mean()
X = X.fillna(X_mean)

# log 변환
# y_log = np.log1p(y)
# y_log_mean = y_log.mean()
# y_log = y_log.fillna(y_log_mean)

# 하이퍼파라미터 탐색 범위
param_grid = {
    'max_depth': [5, 10, 20, None],
    'min_samples_leaf': [1, 3, 5],
    'max_leaf_nodes': [None, 50, 100, 200]
}

# 모델 정의
model = RandomForestRegressor(random_state=777)

# GridSearchCV 설정
cv = KFold(n_splits=3, shuffle=True, random_state=777)
grid_search = GridSearchCV(
    model,
    param_grid=param_grid,
    scoring={
        'rmse': 'neg_root_mean_squared_error',
        'r2': 'r2'
    },
    refit='rmse',  # rmse 기준으로 best_estimator_ 선택
    cv=cv,
    n_jobs=-1,
    verbose=1
)

# grid_search.fit(X, y_log)
grid_search.fit(X, y)

# 결과 출력
print("최적 하이퍼파라미터:", grid_search.best_params_)

# 평균 점수 추출
mean_rmse = -grid_search.cv_results_['mean_test_rmse'][grid_search.best_index_]
mean_r2 = grid_search.cv_results_['mean_test_r2'][grid_search.best_index_]

print("최적 평균 RMSE (log scale):", mean_rmse)
print("최적 평균 R² (log scale):", mean_r2)
```
