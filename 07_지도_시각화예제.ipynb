아파트 실거래 데이터 적용 #1

```python
df = pd.read_csv('/content/drive/MyDrive/kdt_2506/data/trade_apt_api_2023_address.txt', sep='\t')
df['거래금액'] = df['거래금액'].str.replace(',', '').str.replace(' ', '').astype(int)
df
```

```python
df['주소'] = '서울시 ' + df['지역명'] + ' ' + df['법정동'] + ' ' + df['지번']
df_jr = df.loc[(df['지역명']=='종로구')&(df['기준년월']>=202301),['주소', '거래금액']]
df_jr
```

```python
df_jr.info()
```

```python
df_jr_grp = df_jr.groupby('주소').mean()
df_jr_grp
```

위도, 경도 가져오는 함수 만들기 : 더 정확한 위도, 경도는 google, naver, kakao등의 지도 API를 이용

```python
# https://wonhwa.tistory.com/29

from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')

def geocoding(address):
    try:
        geo = geo_local.geocode(address)
        x_y = [geo.latitude, geo.longitude]
        return x_y

    except:
        return [0,0]
```

```python
lat_start, lng_start = geocoding('서울특별시 종로구 종로1길 36')
lat_start, lng_start
```

구한 위도, 경도를 이용하여 지도 시각화

```python
m = folium.Map(location=[lat_start, lng_start], zoom_start=14, tiles='OpenStreetMap', width=800, height=450)

for address, price in zip(df_jr_grp.index, df_jr_grp['거래금액']):
    lat, lng = geocoding(address)
    folium.Marker([lat, lng],
                tooltip=price,
                icon=folium.Icon(color='red', icon='glyphicon-ok')).add_to(m)
display(m)
```
