```python
def get_html(output_file_name, region_name, region_code, this_ym):
    headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.89 Safari/537.36'}
    page_url = "http://apis.data.go.kr/1613000/RTMSDataSvcAptTrade/getRTMSDataSvcAptTrade?LAWD_CD="+str(region_code)+"&DEAL_YMD="+str(this_ym)+"&serviceKey="+secret_key
    print(page_url)

    response = requests.get(page_url, headers=headers, verify=False)
    tree = ElementTree.fromstring(response.content)

    elements = tree.iter(tag="item")

    for element in elements:
        price = element.find("dealAmount").text
        const_year = element.find("buildYear").text
        year = element.find("dealYear").text
        month = element.find("dealMonth").text
        day = element.find("dealDay").text
        dong = element.find("umdNm").text
        apt_name = element.find("aptNm").text
        square = element.find("excluUseAr").text
        stair = element.find("floor").text

        elm_list = [this_ym, region_name, region_code, dong, apt_name, price, year, month, day, const_year, square, stair]
        print(elm_list)

        output_file = open(output_file_name, "a", encoding="utf-8")
        output_file.write("{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\n".format(this_ym, region_name, region_code, dong, apt_name, price, year, month, day, const_year, square, stair))
        output_file.close()

    return
```

```python
def main():
    for this_ym in yms:
        get_html(output_file_name, region_name, region_code, this_ym)

    return
```

```python
main()
```
