```python
df_num = df.select_dtypes(include=np.number)
df_num.head()
```

*  가장 많이 치우쳐있는 column이름과 왜도값을 출력

```python
df_num.skew()
```

```python
df_num.skew().sort_values(ascending=False).head(1)
```

*  가장 많이 몰려있는 column이름과 첨도값, 분산, 표준편차를 출력

```python
df_kurt = df_num.kurt().sort_values(ascending=False)
df_kurt
name = df_kurt.index[0]
krut = df_kurt[0]
```

```python
var = df_num.var()['전용면적']
```

```python
std = df_num.std()['전용면적']
```

```python
name, krut, var, std
```

```python
print("Column {}의 첨도가 {:.2f}로 가장 높고, 분산은 {:.2f}, 표준편차는 {:.2f}입니다.".format(name, krut, var, std))
```

*  동별 거래금액 평균을 구하여 출력

```python
df.columns
```

```python
df_dong = df.drop(['지역명', '아파트'], axis=1)
df_dong
```

```python
pd.options.display.float_format = '{:.2f}'.format
df_dong.groupby('법정동').mean()['거래금액']
```

*  일별 거래금액 추이를 출력

```python
df_day = df.groupby(['일']).sum()['거래금액']
df_day
```

*  평당단가가 가장 높았던 일과 동을 구하여 출력(예:2020.06.01 내수동)

```python
df['일'].unique()
```

```python
df['법정동'].unique()
```

```python
df_grp = df.groupby(['일', '법정동']).sum()
df_grp.head(3)
```

```python
df_grp = df.groupby(['일', '법정동'], as_index=False).sum()
df_grp.head(3)
```

```python
df_grp['평단가'] = df_grp['거래금액'] / (df_grp['전용면적']/3.3)
df_grp.sort_values('평단가', ascending=False)
```

```python
df_grp.sort_values('평단가', ascending=False).head(1)[['일', '법정동', '거래금액', '전용면적', '평단가']]
# df_grp.sort_values('평단가', ascending=False).reset_index(drop=True).head(1)[['일', '법정동', '거래금액', '전용면적', '평단가']]
```

*  거래금액과 가장 상관관계가 높은 column의 순서대로 이름과 상관계수를 출력
